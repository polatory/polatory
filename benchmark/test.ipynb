```r
library(gstat)
library(sp)
```

```r
data <- data.frame(x = runif(50000, min = 0, max = 1000),
                   y = runif(50000, min = 0, max = 1000),
                   z = runif(50000, min = 0, max = 1000))
```

```r
g <- gstat(formula = value ~ 1, locations = ~ x + y + z,
           model = vgm(1.0, "Sph", 50.0, anis = c(135, 45, 0, 0.7, 0.3),
                       add.to = vgm(1.0, "Sph", 150.0, anis = c(135, 45, 0, 0.6, 0.4), nugget = 0.1)),
           beta = 0.0, nmax = 30, dummy = TRUE)
```

```r
sim <- predict(g, newdata = data, nsim = 1)
write.csv(sim, "sim.csv", row.names = FALSE)
```
